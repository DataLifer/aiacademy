### 2. 음성
1. 기술설명
* 네이버 클로바(인공지능 음성지원 플랫폼)
  * 스트리밍형식의 머신러닝기반 음악추천 및 재생, 영어 프리토킹 등 음성인식기능 및 음성합성기능 제공

2. 개발 CONTENTS 
  1. Clova Speech Recognition(클로바 음성인식) : 인간의 목소리 인식하여 작동하는 비서 애플리케이션
  2. Clova Speech Synthesis(클로바 음성합성) : 텍스트를 성우의 음성으로 자연스럽게 읽어주는 API
  3. Clova Premium Voice(클로바 프리미엄 음성합성) : 일반 음성합성 + 감정반영 
  4. Clova Extension(클로바 익스텐션 : 생활, 학습, 금융, 쇼핑 부가서비스제공) : 검색뿐 아닌, 생활, 학습, 금융, 쇼핑 등 다양한 분야의 서비스를 클로바를 통해 제공

3. 주요성과
  1. 검색, 날씨, 음악추천 및 재생, 번역, 프리토킹 등의 다양한 기능 제공을 통한 사용자비서 및 편의성 효율성 증대
  2. 음성학습기술을 통한 음성안내시스템, 뉴스, E-BOOK 등의 글읽기 애플리케이션에 필요한 서비스제공 
  3. 음성을 통한 가전제품 및 공동주택의 홈네트워크 제어, 배달주문, 금융서비스에 적용가능
  4. 한국어, 영어, 일본어, 중국어 등에서 인식가능하며 국내최고의 뛰어난 한국어인식률 보유
  5. 음성인식에 슬픔이나 기쁨등의 사용자감정을 반영하여 예체능관련서비스적용이 가능
