```
import requests # 요청라이브러리 임포트

url = "https://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=50ee936c7380f8675b6bb4de9a74212e&targetDt=20120101" # 가져올 url주소 

res = requests.get(url) # get방식으로 응답객체 res 생성

text = res.text # 응답객체 text 생성
MD_json = json.loads(text) # MD = MOVIE DATA를 json방식으로 불러옴

```

```
print(json.dumps(MD_json, indent = 4, sort_keys = True))
```

```
print(MD_json.keys()) # 어떠한 키를 사용하는 지 확인

print(MD_json['boxOfficeResult'].keys()) # 3가지의 키중 앞서 확인한 'boxOfficeResult'정보의 구조확인
```

```
for i in MD_json['boxOfficeResult']['dailyBoxOfficeList']: # MD_json 객체의 'boxOfficeResult'키에서 'dailyBoxOfficeList을 리스트순서대로 반복
  print(i['rank'], i['rankOldAndNew'], i['movieCd'], i['movieNm'], i['salesAmt'])   # rank, rankOldAndNew, movieCd, movieNum, saleAmt 출력
```

```
import pandas as pd # pandas 라이브러리 임포트 알리아스 pd

movie = [] # 빈 리스트를 변수 movie에 생성

for i in MD_json['boxOfficeResult']['dailyBoxOfficeList']: # MD_json 객체의 'boxOfficeResult'키에서 'dailyBoxOfficeList을 리스트순서대로 반복
  movie.append([i['rank'], i['rankOldAndNew'], i['movieCd'], i['movieNm'], i['salesAmt']]) # rank, rankOldAndNew, movieCd, movieNum, saleAmt를 빈 movie 리스트에 삽입



data = pd.DataFrame(movie) # 채워진 movie를 Dataframe으로 생성

data # 출력
```

2. 판매처별 공적 마스크 판매처 정보 API 활용하기

```
import json # json 라이브러리 임포트

import requests # requests 라이브러리 임포트

url2 = "https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/sales/json?page=1&perPage=500" # 요청할 주소(String 타입)를 변수 url2에 대입

res = requests.get(url2) # 요청한 url2에 대한 응답객체 생성

text = res.text # 응답객체 텍스트 생성

corona19_mask = json.loads(text)  # 응답객체 텍스트를 json으로 불러온 뒤 변수 corona19_mask에 대입

print(json.dumps(corona19_mask, indent = 4, sort_keys = True)) # 변수 corona19_mask를 덤프함수 적용( 들여쓰기 4, 정렬을 실시)후 출력


```

```
corona19_mask_list = [] # 빈 리스트를 corona19_mask_list에 생성

for i in corona19_mask['sales']: # corona19_mask 객체의 'sales' 반복
  corona19_mask_list.append([i['created_at'], i['code'], i['remain_stat']]) # 리스트안에 존재하는 create_at, code, remain_stat를 corona19_mask_list에 삽입

corona19_mask_dataframe = pd.DataFrame(corona19_mask_list) # 삽입된 corona19_mask_list를 데이터프레임으로 저장


corona19_mask_dataframe.columns = ['기준시간', '마스크현황', '식별코드'] # 컬럼명을 기준시간, 마스크현황, 식별코드로 변경
corona19_mask_dataframe.head(10) # 위에서 10번째까지 출력
```
